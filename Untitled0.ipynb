```python
import pandas as pd

# Function to calculate 5-point summary and outliers for a given region
def calculate_summary_and_outliers(region_data):
    q1 = region_data.quantile(0.25)
    q3 = region_data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Get outliers
    outliers = region_data[(region_data < lower_bound) | (region_data > upper_bound)]

    # Prepare a dictionary with summary information
    summary_dict = {
        'Q1': q1,
        'Q3': q3,
        'IQR': iqr,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outliers': outliers.to_dict()
    }

    return summary_dict

# Group data by world region and calculate the 5-point summary for each region
world_region_summary = {}

# Set to store unique outliers across all world regions
unique_outliers = set()

for world_region, region_data in mpi_subnational.groupby('World region')['MPI Regional']:
    world_region_summary[world_region] = calculate_summary_and_outliers(region_data)

    # Get unique outliers for this world region and add to the set
    unique_outliers.update(set(world_region_summary[world_region]['Outliers'].items()))

# Print the 5-point summary and outliers for each world region
for world_region, summary in world_region_summary.items():
    print(f"World Region: {world_region}")
    print(f"5-Point Summary:")
    for key, value in summary.items():
        print(f"{key}: {value}")

    print("\nOutliers:")
    # Filter unique outliers for this world region
    region_outliers = {index: mpi for index, mpi in unique_outliers if index in summary['Outliers'].keys()}
    for index, mpi in region_outliers.items():
      country_name = mpi_subnational.loc[index, 'Country']
      regional_name = mpi_subnational.loc[index, 'Sub-national region']
      print(f"{country_name}: {regional_name} : MPI {mpi:.3f}")

    print("\n" + "="*30 + "\n")

```

```python
from scipy.stats import spearmanr

# Selecting the relevant columns
features = ["Nutrition", "Years of Schooling", "School Attendance", "Child Mortality", "Cooking Fuel", "Sanitation", "Drinking Water", "Electricity", "Housing", "Assets"]

# Creating an empty dictionary to store correlations
correlations = {}

# Calculating Spearman rank correlation for each feature with MPI
for feature in features:
    # Handling NaN values if any
    mask = ~np.isnan(data[feature])

    # Calculating Spearman rank correlation
    correlation, _ = spearmanr(data['MPI'][mask], data[feature][mask])

    # Storing the correlation coefficient
    correlations[feature] = correlation

# Displaying all the correlation coefficients
print("Correlation Coefficients:")
for feature, correlation in correlations.items():
    print(f'{feature}: {correlation:.3f}')

# Sorting features based on their correlation with MPI in descending order
sorted_features = sorted(correlations, key=correlations.get, reverse=True)

# Selecting the top 3 features
top_3_features = sorted_features[:3]

# Displaying the selected top 3 features
print("\nTop 3 Features:")
for feature in top_3_features:
    print(f'{feature}: {correlations[feature]:.3f}')
```

```python
#wilcoxon signed-rank test to check if there is a significant difference between population falling in multidimention poverty
from scipy.stats import wilcoxon

# Assuming 'MPI Urban' and 'MPI Rural' are the columns representing poverty in urban and rural regions
urban_mpi = mpi_national['MPI Urban'].dropna()
rural_mpi = mpi_national['MPI Rural'].dropna()

# Perform Wilcoxon signed-rank test
statistic, p_value = wilcoxon(urban_mpi, rural_mpi)

# Print the results
print(f"Test Statistic: {statistic}")
print(f"P-Value: {p_value}")

# Set significance level (e.g., 0.05)
alpha = 0.05

# Check for significance
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between urban and rural MPI values.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between urban and rural MPI values.")
```
